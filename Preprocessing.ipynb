```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import re
import Sastrawi 
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
```

```python
youtube_comment = pd.read_excel('UASLucknut2.xlsx')
print(youtube_comment)
```

```python
youtube_comment.shape
```

```python
youtube_comment = youtube_comment.drop_duplicates(subset=['Komentar'])
```

```python
youtube_comment.isnull().sum()
```

```python
youtube_comment= youtube_comment.dropna()
```

```python
youtube_comment.shape
```

```python
def clean_youtube_comment(text):
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'RT[\s]+', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'[^A-Za-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
youtube_comment['Komentar'] = youtube_comment['Komentar'].apply(clean_youtube_comment)
```

```python
youtube_comment['Komentar'] = youtube_comment['Komentar'].str.lower()
```

```python
youtube_comment
```

```python
#normalisasi
norm= {"yg":"yang","dr":"dari","sy":"saya","gak":"tidak","bool":"bokong","bokep":"pornografi"}

def normalisasi(str_text):
    for i in norm:
        str_text=str_text.replace(i, norm[i])
    return str_text

youtube_comment['Komentar'] = youtube_comment['Komentar'].apply(lambda x: normalisasi(x))
youtube_comment
```

```python
#Stopwords
more_stop_words = ["yang", "dan", "di","ke","dari","untuk","pada","dengan","adalah","itu", "ini", "dalam", "tidak", "oleh", "sebagai", "namun", "tetapi", "karena", "atau", "jika", "bahwa", "saja", "bisa", "sangat", "juga", "hanya", "sudah", "saya", "kamu", "kita", "mereka", "ia", "nya", "apa", "bagaimana","ketika", "setelah", "sebelum", "semua", "lebih", "sudah", "sehingga"]

stop_words = StopWordRemoverFactory().get_stop_words()
stop_words.extend(more_stop_words)

new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stopword(str_text):
  str_text = stop_words_remover_new.remove(str_text)
  return str_text

youtube_comment['Komentar'] = youtube_comment['Komentar'].apply(lambda x: stopword(x))
youtube_comment
```

```python
# Tokenize
tokenized = youtube_comment['Komentar'].apply(lambda x:x.split())
tokenized
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(text_cleaning):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in text_cleaning:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  print(d_clean)
  return d_clean

tokenized = tokenized.apply(stemming)

tokenized.to_csv("Komen_Preprocessing2.csv",index=False)
```
